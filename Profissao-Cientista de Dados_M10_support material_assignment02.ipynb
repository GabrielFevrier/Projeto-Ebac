````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

np.random.seed(28)

df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

### Observações a cerca dos dados  
Numa primeira análise, as variáveis relacionadas a veículos mostraram-se boas preditoras de renda.   
Entretanto, ao introduzir a separação por gênero, os gráficos indicam que a variável "posse_de_veículo" não seria uma preditora tão robusta,  
uma vez que seus intervalos de confiança se sobrepõem em ambos os sexos.

````python
imovel = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, ci = 90)
````

````python
veiculo = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, ci = 90)
````

````python
imovel_por_sexo = imovel = sns.pointplot(x="posse_de_imovel", y="renda", hue = 'sexo',
                   data=df, dodge=True, ci = 90)
````
